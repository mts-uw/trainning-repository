```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston = load_boston()
feat = pd.DataFrame(boston.data, columns=boston.feature_names)
target = boston.target
feat.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor
```

```python
feat_trn, feat_tes, target_trn, target_tes = train_test_split(feat, target)
feat_trn
```

```python
model = ExtraTreesRegressor(n_estimators = 500)
model.fit(feat_trn, target_trn)
```

```python
target_pred =  model.predict(feat_tes)
```

```python
from sklearn.metrics import mean_squared_error
import math
```

```python
print(math.sqrt(mean_squared_error(target_pred, target_tes)))
```

```python
import matplotlib.pyplot as plt
plt.hist(target, bins =30)
```

```python
import shap

model.fit(feat, target)

shap.initjs()

explainer = shap.TreeExplainer(model=model)
```

```python
shap_values = explainer.shap_values(feat)
```

```python
shap.summary_plot(shap_values, feat)
```

```python
shap.force_plot(base_value=explainer.expected_value, shap_values=shap_values, features=feat)
```
